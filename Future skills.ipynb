```python
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['skill'])
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim)
```

```python
cosine_sim_df
```

```python
def content_recommendation(title):
    a = df.copy().reset_index().drop('index',axis=1)
    index = a[a['skill']==title].index[0]
    top_n_index = list(cosine_sim_df[index].nlargest(20).index)
    try:
        top_n_index.remove(index)
    except:
        pass
    similar_df = a.iloc[top_n_index][['title']]
    similar_df['similarity'] = cosine_sim_df[index].iloc[top_n_index]
    return similar_df
```

```python
title = 'etl,sql,apache-spark,java,'
s=content_recommendation(title)
```

```python
s  = s.to_json()
s
```

```python
t = list(s["title"])
si = list(s["similarity"])
o = []
t1=[]
t2=[]

for i in t:
    d  = {'job':""}
    d["job"] = i
    t1.append(d)
t1
for j in si:
    k = {"similarity":""}
    k["similarity"] = j
    t2.append(k)
for i in range(len(t1)):
    t1[i].update(t2[i])
    o.append(t1[i])
o    
    
    
```
